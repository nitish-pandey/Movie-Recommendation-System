# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import re
import ast
```

# Data Loading ( Movies MetaData )

```python
metadata=pd.read_csv("movies_metadata.csv")
metadata.head()
```

```python
print(metadata.columns,'\n\n',metadata.shape)
```

```python
meta_data=metadata.drop(['belongs_to_collection','homepage','budget','original_title','production_companies','production_countries','popularity','poster_path','revenue','spoken_languages','status','video','vote_average','vote_count'],axis=1)
meta_data.head()
```

```python
def generate_genre(data):
    genre=[]
    for i in data:
        a=i.get('name')
        if a=='Science Fiction':
            a='scifi'
        genre.append(a)
    if genre==[]:
        return np.NaN
    return " ".join(genre)
```

```python
meta_data['genres'] = meta_data['genres'].map(lambda x: ast.literal_eval(x))
meta_data['genres'] = meta_data['genres'].map(lambda x: generate_genre(x))
meta_data.head()
```

```python
meta_data.to_csv('final_metadata.csv',index=False)
```

# Data Loading 2 ( Credits )

```python
credits=pd.read_csv("credits.csv")
credits.head()
```

```python
def generate_cast1(data):
    cast=[]
    for i in data:
        a=i.get('name')
        cast.append(a)
    if cast==[]:
        return np.NaN
    return cast[0]

def generate_cast2(data):
    cast=[]
    for i in data:
        a=i.get('name')
        cast.append(a)
    if cast==[] or len(cast)<=1:
        return np.NaN
    return cast[1]

def generate_cast3(data):
    cast=[]
    for i in data:
        a=i.get('name')
        cast.append(a)
    if cast==[] or len(cast)<=2:
        return np.NaN
    return cast[2]

def generate_director(data):
    dir=[]
    for i in data:
        if i.get('job')=='Director':
            dir.append(i.get('name'))
    if dir==[]:
        return np.NaN
    return (" ".join(dir))
```

```python
credits['cast'] =credits['cast'].map(lambda x: ast.literal_eval(x))
credits['crew'] = credits['crew'].map(lambda x: ast.literal_eval(x))
```

```python
credits['actor_1_name'] = credits['cast'].map(lambda x: generate_cast1(x))
credits['actor_2_name'] = credits['cast'].map(lambda x: generate_cast2(x))
credits['actor_3_name'] = credits['cast'].map(lambda x: generate_cast3(x))

credits['director_name'] =credits['crew'].map(lambda x: generate_director(x))
```

```python
credits.drop(['cast','crew'],axis=1,inplace=True)
credits.head()
```

```python
credits.to_csv('final_credits.csv',index=False)
```
