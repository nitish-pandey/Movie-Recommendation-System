# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import re
import ast
```

# Data Loading and Merging

```python
metadata=pd.read_csv('/content/drive/MyDrive/My Codes and data/Movie-Data/final_metadata.csv')
metadata.head()
```

```python
credits=pd.read_csv('/content/drive/MyDrive/My Codes and data/Movie-Data/final_credits.csv')
credits.head()
```

```python
metadata=metadata[metadata["title"].str.contains("NaN")==False]

metadata['id']=metadata['id'].astype(int)
credits['id']=credits['id'].astype(int)

merge=pd.merge(metadata,credits,on='id')
```

```python
merge
```

```python
data=merge
data=data.replace(to_replace=np.NaN,value="")
data['features']=data['features']=data['title']+" "+data['director_name']+" "+data['genres']+" "+data['actor_1_name']+" "+data['actor_2_name']+" "+data['actor_2_name']
data.features=data.features.str.lower()
data.features=data.features.replace('unknown',' ')
data=data.replace('\xa0','',regex=True).astype(str)
```

```python
data
```

```python
merge=data[['id','features']]
merge
```

```python
merge.to_csv('features.csv',index=False)
```
